# Data Preparation

# Prediction

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, plot_confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

## Predicting the zodiac sign

### Data Preparation

```python
cols1 = ['body_type', 'diet', 'orientation', 'pets', 'religionCleaned',
       'sex', 'job', 'signsCleaned']
df1 = profiles[cols1].dropna()
```

```python
for col in cols1[:-1]:
    df1 = pd.get_dummies(df1, columns=[col], prefix=[col])
```

```python
df1.head()
```

```python
#y is the target column, X has the rest
X = df1.iloc[:, 1:len(df1.columns)]
y = df1.iloc[:, 0:1]

#Split the data into chunks
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

#turn into 1d arrays
y_train = y_train.to_numpy().ravel()
y_test = y_test.to_numpy().ravel()
```

### Logistic Regression

#### Training

```python
lr = LogisticRegression(multi_class="multinomial")
lr.fit(X_train, y_train)
```

#### Evaluation

```python
lr_predictions = lr.predict(X_test)
```

```python
print(classification_report(y_test, lr_predictions))
```

### K-nearest neighbors

#### Training

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

#### Evaluation

```python
knn_predictions = knn.predict(X_test)
```

```python
print(classification_report(y_test, knn_predictions))
```

### Decision Tree

#### Training

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train) 
```

#### Evaluation

```python
tree_predictions = tree.predict(X_test)
```

```python
print(classification_report(y_test, tree_predictions))
```

```python
tree_cm = confusion_matrix(y_test, tree_predictions)
tree_labels = tree.classes_
```

```python
plt.figure(figsize=(10,7))

ax = plt.subplot()
sns.heatmap(tree_cm, annot=True, ax=ax, fmt="d")

# labels, title and ticks
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels') 
ax.set_title('Confusion Matrix')
ax.yaxis.set_tick_params(rotation=360)
ax.xaxis.set_tick_params(rotation=90)

ax.xaxis.set_ticklabels(tree_labels)
ax.yaxis.set_ticklabels(tree_labels);
```

The results of the evaluation show that the accuracy of the classification of the zodiac signs is very low. The best prediction was made by the K-nearest-neighbor classifier. It achieved an accuracy score of 9 %. The chance of guessing the sign right is 8.3 % which is the baseline. The confusion matrix also shows that the classification does not function very well, because there is no pattern recognizable. Ideally there should be an diagonal with large values, that represent the true positives. One can conclude that there is no relation between the sign of a person and the chosen features.

## Predicting the zodiac sign with essays

### Data Preparation

```python
profiles['all_essays'] = profiles.essay0 + profiles.essay1 + profiles.essay2 + profiles.essay3 + profiles.essay4 + profiles.essay5 + profiles.essay6 + profiles.essay7 + profiles.essay8 + profiles.essay9
```

```python
df2 = profiles.dropna(subset=['all_essays', 'signsCleaned'])
```

```python
labels = df2['signsCleaned']
corpus = df2['all_essays'].tolist()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
counter = CountVectorizer()
counts = counter.fit_transform(corpus)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
labels = le.fit_transform(labels)
```

```python
train_counts, test_counts, train_labels, test_labels = train_test_split(counts, labels, test_size=0.2)
```

### Random Forest

```python
forest = RandomForestClassifier()
forest.fit(train_counts, train_labels)
forest_predictions = forest.predict(test_counts)
```

```python
print(classification_report(test_labels, forest_predictions))
```

The Random Forest classifier makes a bad prediction with an accuracy score of 10 %, which is only a little bit higher than the baseline of 8.3 %.
